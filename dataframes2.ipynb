```python
dfProduct = pd.read_csv('products.csv')
dfProduct = dfProduct.applymap(lambda x: x.strip() if isinstance(x,str) else x)
dfProduct.head()
```

```python
#List all the products
for product in dfProduct.Product.unique():
    print(product,"\n")
```

```python
# List all country names
for product in dfProduct["Retailer country"].unique():
    print(product,"\n")
```

```python
# List all order methods
for product in dfProduct["Order method type"].unique():
    print(product,"\n")
```

```python
# List the product which is able to sell maximum numbers in United States

countries = dfProduct.groupby('Retailer country')
usaData = countries.get_group('United States') 
usaData[usaData["Quantity"] == usaData["Quantity"].max()][["Retailer country","Product","Quantity"]]
```

```python
# Find the order method type which is able to sell more number of products in Germany
germanyData = countries.get_group('Germany')
orderType = germanyData.groupby('Order method type').sum()
dfres = orderType.sort_values("Quantity",ascending = False)
dfres[["Quantity"]].head(1)
# germanyData[germanyData["Quantity"] == germanyData["Quantity"].max()][["Retailer country","Product","Order method type","Quantity"]]
```

```python
# List the top 5 products which has more gross profit in Australia

australiaData = countries.get_group('Australia')
grossProfit = australiaData.groupby("Product")["Gross profit"].sum()
grossProfit.sort_values(ascending = False).head()
```

```python
# Which order method is more popular (based on Quantity) world wide?
orderType = dfProduct.groupby('Order method type').sum()
dfres = orderType.sort_values("Quantity",ascending = False)
dfres[["Quantity"]].head(1)
```

```python
# List the top and bottom three products sold in Canada in year 2006

CanadaData = countries.get_group('Canada')
dfresgot = CanadaData.loc[CanadaData['Year'] == 2006]
array = dfresgot["Product"].unique()
print(array[:3])
print(array[-3:])
```
